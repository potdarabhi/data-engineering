```sql
SELECT DISTINCT * FROM sales
```

```sql
SELECT c.full_name, s.product,s.region_name
from cust c 
LEFT join sales s 
on s.customer_id = c.customer_id
```

```sql
select c.customer_id, c.full_name, count(s.product) as quantity
from cust c
left join sales s 
on c.customer_id = s.customer_id
group by 1,2
order by 1
```

```sql
select distinct * 
from cust c
left join sales s
on c.customer_id = s.customer_id
where c.gender = 'F'
and s.amount > 500
```

```sql
select s.region_name, sum(s.amount) as total_amount
from cust c
join sales s
on c.customer_id = s.customer_id
group by 1
order by 1
```
