```sql
CREATE TABLE emp1 (emp_id INT,emp_name STRING,sale_amount INT,dept STRING,sale_date DATE);
```

```sql
SELECT * FROM emp1
```

```sql
SELECT e.emp_id,e.emp_name,e.dept,e.sale_amount,
rank() over(partition by e.dept order by e.sale_amount desc) as rank
FROM emp1 e
```

```sql
SELECT e.emp_id,e.emp_name,e.sale_amount,
sum(e.sale_amount) over(partition by e.emp_id order by e.emp_id) as running_total
from emp1 e
```
