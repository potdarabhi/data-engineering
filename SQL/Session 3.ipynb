```sql
SELECT c.first_name,c.last_name,od.total_amount FROM customers c 
INNER JOIN orders od ON c.customer_id = od.customer_id
WHERE od.total_amount >
(SELECT AVG(o.total_amount) FROM orders o)
```

```sql
--select * from order_details
SELECT od.product_name, sum(quantity) as total_quantity
FROM order_details od GROUP BY od.product_name HAVING sum(quantity) > 2
```

```sql
with customers_orders as(
  select c.first_name, c.last_name,sum(od.quantity) as total_quantity
  from customers c
  inner join orders o on c.customer_id = o.customer_id
  inner join order_details od on o.order_id = od.order_id
  group by c.first_name,c.last_name
  --having sum(od.quantity) > 2
)
select * from customers_orders
```

```sql
with customer_orders as(
  select od.product_name,sum(od.price_each * od.quantity) as total_revenue
  from order_details od 
  group by 1
)

select * from customer_orders where total_revenue > 5000
```

## Case statement

```sql
with product_categorized as (
select o.customer_id,o.order_id,o.total_amount,
case
  when o.total_amount < 1000 then 'Small'
  when o.total_amount BETWEEN 1000 AND 2500 THEN 'Medium'
  else 'Large'
end as size
from orders o
)







select * from product_categorized where size = 'Large'
```

```sql
select c.first_name, c.last_name,count(o.order_id) as total_orders,
case 
  when total_orders = 1 then 'New'
  when total_orders > 1 then 'Repeat'
  else 'Yet to order'
end as customer_type
from customers c
left join orders o on c.customer_id = o.customer_id
group by c.first_name,c.last_name
```

## Function (String + date)

```sql
select upper(c.first_name) as Name from customers c 
```

```sql
select upper(concat(c.first_name," ",c.last_name)) as full_name from customers c
```

```sql
select month(o.order_date) as order_month, count(o.order_id) as total_orders 
from orders o
group by order_month
order by order_month

```

```sql
with customers_details as(
  select c.customer_id,concat(c.first_name ," ",c.last_name) as full_name,sum(od.price_each * od.quantity) as total_spend
  from customers c
  inner join orders o on c.customer_id = o.customer_id
  inner join order_details od on o.order_id = od.order_id
  group by c.customer_id,c.first_name,c.last_name
)

select full_name, total_spend,
case
  when total_spend < 10000 then 'Bronze'
  when total_spend BETWEEN 10001 and 25000 then 'Silver'
  else 'Gold'
end as customer_category
from customers_details cd 
left join order_details od
on cd.customer_id = od.customer_id
```
