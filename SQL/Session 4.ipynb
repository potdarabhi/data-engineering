```sql
select
order_id,
order_date,
row_number() over(partition by order_id order by order_date) as rn
from orders
```

```sql
--Ranks rows by quantity; ties get same rank, but gaps are left in the sequence.
select od.product_name, od.quantity,
rank() over(order by od.quantity desc) as ranking
from order_details od
```

```sql
select od.order_id,
od.product_name,
od.quantity,
rank() over(partition by od.order_id order by od.quantity desc) as ranking
from order_details od
```

```sql
SELECT 
od.product_name,od.quantity,
dense_rank() over(order by od.quantity desc) as DenseRank,
rank() over(order by od.quantity desc) as Rank
from order_details od
```

```sql
SELECT 
od.product_name,
SUM(od.quantity) as quantity,
dense_rank() over(order by sum(od.quantity) desc) as popularity_rank
FROM order_details od
Group By 1
```

```sql
--LAG() - COMPARE WITH PREVIOUS ROWS

SELECT 
o.order_id,
o.order_date,
LAG(o.order_date) over(order by o.order_date) as previous_date
FROM orders o
```

```sql
--LEAD() - COMPARE WITH NEXT ROWS

SELECT 
o.order_id,
o.order_date,
LEAD(o.order_date) over(order by o.order_date) as next_date
FROM orders o
```

```sql
SELECT
c.first_name,
o.order_id,
od.quantity,
lag(od.quantity) over(partition by c.customer_id order by od.order_id) as previous_quantity
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_details od ON o.order_id = od.order_id
```

```sql
SELECT
c.first_name,
od.product_name,
lead(od.product_name) over(partition by c.customer_id order by od.product_name) as next_product
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_details od ON o.order_id = od.order_id
```

```sql
--SUM() ---> RUNNING TOTAL OR Cummulative Sum 

SELECT
o.customer_id,
o.total_amount as amount,
sum(o.total_amount) over(PARTITION BY o.customer_id ORDER BY o.order_id) as running_total
FROM orders o 
```

```sql
SELECT first_name,
department,
salary,
sum(salary) OVER(PARTITION BY department ORDER BY emp_id,department) as running_total
FROM employees
```
