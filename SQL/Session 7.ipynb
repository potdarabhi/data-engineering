**SELECT * V/S SELECT Columns**

```sql
SELECT * 
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_details od ON o.order_id = od.order_id;
```

```sql
SELECT c.first_name, o.order_id, od.product_name, od.price_each
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_details od ON o.order_id = od.order_id;
```

### Filter After JOIN V/S Filter Before

```sql
--❌ Unoptimized:

SELECT c.first_name, od.product_name, od.price_each
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_details od ON o.order_id = od.order_id
ORDER BY od.price_each DESC;
```

```sql

--Optimised

SELECT c.first_name, od.product_name, od.price_each
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_details od ON o.order_id = od.order_id
WHERE od.price_each > 1000
ORDER BY od.price_each DESC;
```

### GROUP BY without filtering vs with early filtering

```sql
--Unoptmised query
SELECT c.customer_id, SUM(od.price_each * od.quantity) AS total_spent
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_details od ON o.order_id = od.order_id
GROUP BY c.customer_id;
```

```sql
-- Optimized:

SELECT c.customer_id, SUM(od.price_each * od.quantity) AS total_spent
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_details od ON o.order_id = od.order_id
WHERE o.order_date >= '2023-01-01'
GROUP BY c.customer_id;
```
