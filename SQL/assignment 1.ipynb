```sql
create table sales(
  sale_id int,
  customer_id	int,
  product	varchar(50),
  amount decimal(10,2),
  sale_date date,
  region_name varchar(50)
);
```

```sql
insert into sales values
(101,001,'Shoes',2999,'2023-10-01','North'),
(102,002,'T-Shirt',999,'2023-10-01','West'),
(103,001,'Watch',5999,'2023-10-02','North'),
(104,003,'Bag',1999,'2023-10-03','South');
```

```sql
-- sql query to get total sales per region

select region_name,sum(amount) from sales group by region_name order by sum(amount) desc
; 
```

```sql
-- select data for orders places after 2023-10-02

select distinct * from sales s where s.sale_date >= '2023-10-02'
```

```sql
-- select data for where the order amount is greater than 2000

select distinct * from sales s where s.amount > 2000
```

```sql
-- select data for orders where the amount is max based on region 

select distinct region_name, max(amount) from sales group by region_name;
-- select distinct * from sales
```
