```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("DataX").getOrCreate()
```

```python
# create realistic DataFrame

data = {
    ('Alice','IT',50000),
    ('Bob','HR',60000),
    ('Charlie','Operations',70000),
    ('David','Engineering',80000),
    ('Eve','HR',90000),
    ('Frank','Finance',10000),
    ('Grace','Accounts',110000),
    ('Hannah','HR',120000),
    ('Ivan','Finance',130000)
}

column = ['Name','Department','Salary']

df = spark.createDataFrame(data, column)
display(df)
```

```python
#Select and filter on DataFrame

display(df.select('Name','Salary'))
```

```python
#filter rows on salary
display(df.filter(df.Salary > 100000))
```

```python
#apply multiple filters

display(df.filter((df['Department']=='Finance') & (df['Salary'] > 100000)))

```

```python
#create a dervied column

from pyspark.sql.functions import col

#add 10% bonus to everyone
df = df.withColumn("Bonus", col("Salary") * 0.1)
display(df)
```

```python
#GroupBy and Aggregations

#avg of salary by departments

display(df.groupBy('Department').avg('Salary'))
```

```python
#apply multiple aggregations

from pyspark.sql.functions import avg, max, min

display(df.groupBy('Department').agg(
    avg('Salary').alias('avg_salary'),
    max('Salary').alias('max_salary'),
    min('Salary').alias('min_salary')
))
```

```python
# Join of DataFrames
#create DataFrame for Dept_Code

dept_data = [('IT',100),('HR',200),('Operations',300),('Engineering',400),('Finance',500),('Accounts',600)]
dept_column = ['Department','Dept_Code']

df_dept = spark.createDataFrame(dept_data, dept_column)

df_joined = df.join(df_dept, on ="Department", how = 'inner')
display(df_joined)
```

```python
#transformation v/s action

#tranformation
filtered = df.filter(df['Salary'] > 80000)

#action
filtered.show()
```

```python
from pyspark.sql.functions import col, avg

data1 = [('Abhijeet','IT',60000),
         ('Rohan','HR',60000),
         ('Pratik','Sales',70000),
         ('Priyanka','Finance',80000)]

columns = ['Name','Department','Salary']

df_task = spark.createDataFrame(data1,columns)

# Task 1: Filter IT employees with salary > 50K
df_task.filter((col('Salary') > 50000) & (col('Department') == 'IT')).show()

# Task 2: Add column "Hike_Amount" as 15% of salary
df_task.withColumn('Hike_Amount', col('Salary') * 0.15).show()

# Task 3: Group by Department and show average salary
df_task.groupBy('Department').avg('Salary').show()
```
