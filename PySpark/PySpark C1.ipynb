```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Spark DataFrames").getOrCreate()
print('Spark session created',spark.version)
```

### Understand Spark DataFrame

```python
#create pyspark dataframe

data = [('Anurag',50000),('Abhijeet',60000),('Pratik',70000),('Aniket',80000)]
column = ['Name','Salary']

spark_df = spark.createDataFrame(data, column)
#spark_df.show()
display(spark_df)
```

```python
import pandas as pd

data1 = [('Anurag',50000),('Abhijeet',60000),('Pratik',70000),('Aniket',80000)]
column1 = ['Name','Salary']

df = pd.DataFrame(data, columns=column1)
display(df)
```

### Real Life Use Case

- Ingesting data from files
- Transforming at scale
- Writing to Delta Tables or Parquet files

```python
#add data type

from pyspark.sql.types import StructType, StructField, StringType, IntegerType

schema = StructType([
    StructField('Name', StringType(), True),
    StructField('Salary', IntegerType(), True)
])

df_typed = spark.createDataFrame(data, schema)
df_typed.show()
df_typed.printSchema()

```
