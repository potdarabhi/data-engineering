- Data Engineers build reusable tools: readers, transformers, loggers.
- OOP helps you modularize, scale, and reorganize pipeline steps cleanly.
- Instead of repeating code for each file format/API — define a class once, use everywhere.


```python
class customer:
  def __init__(self, name, age):
    self.name = name
    self.age = age

c1 = customer('Abhi', 25)
print(c1.name)
```

```python
class Pipelinestep:
  def __init__(self,stepName):
    self.stepName = stepName

  def execute(self):
      print(f'Executing step {self.stepName}')

p1 = Pipelinestep('Ingestion')
p2 = Pipelinestep('Transform')

p1.execute()
p2.execute()
```

### Encapsulation

```python
class Balance:
  def __init__(self):
    self.balance = 1000
    
  def __updateBalance(self, amount):      #private function
    self.balance += amount

  def _showBalance(self):    #protected function
    print(f'Current Balance: {self.balance}')

  def deposit(self,amount):
    if amount > 0:
        self.__updateBalance(amount) #accessing private function
        self._showBalance()   #accessing protected function
    else:
        print('Invalid amount')


dep = Balance()
dep._showBalance()
dep.deposit(500)
```

### Inheritance

```python
class Person:
  def __init__(self,name):
    self.name = name

class Employee(Person):
    def role(self):
        print(f'Employee {self.name} is a Data Engineer')

e1 = Employee('Tejas')
e1.role()
```

```python
class Reader:
    def read(self):
        return "Reading from base..."
    
class CSVReader(Reader):
    def read(self):
        return "Reading from CSV Reader..."
    
class  APIReader(Reader):
    def read(self):
        return "Reading from API Reader..."


#c1 = CSVReader()
#print(c1.read())

print(CSVReader().read())
print(APIReader().read())
```

### Polymorphism

```python
def run_reader(reader):
    print(reader.read())

run_reader(CSVReader())
run_reader(APIReader())
```

###Abstraction

```python
from abc import ABC, abstractmethod
```

```python
class Ingestor(ABC):
    @abstractmethod
    def read(self):
        pass


class CSVIngestor(Ingestor):
    def read(self):
        print("Reading from CSV")


class APIIngestor(Ingestor):
    def read(self):
        print("Reading from API")

def runingestor(ingestor):
    ingestor.read()

runingestor(CSVIngestor())
runingestor(APIIngestor())
```
