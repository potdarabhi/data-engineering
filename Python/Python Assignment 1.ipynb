```python
cust = {'name': 'Pratik','location':'Pune','order_toatal':20000}
type(cust)
```

```python
#list of products
products = ['Shirt','Pants','Shoes','Watch','Perfume']
type(products)

#loop through each product

for pr in products:
    print(pr)
```

```python
#if-else to check order_total

if cust['order_toatal'] > 10000:
    print('High value order')
else:
    print('normal order')
```

```python
#discount function
def discount(info):
    if info['order_toatal'] > 10000:
        return info['order_toatal'] * 0.1
    else:
        return 0
```

```python
#Calculate discounted price and update cust
disc = discount(cust)
cust['discount'] = disc
cust['discounted_price'] = cust['order_toatal'] - disc
print(cust)
```
