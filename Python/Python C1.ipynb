IF-ELSE block


```python
def tax_bracket(salary):
    if salary >= 70000:
        return "30%"
    elif salary >= 50000:
        return "20%"
    else:
        return "10%"
```

```python
salary = int(input('Enter your salary'))
print(f'Your tax bracket is {tax_bracket(salary)}')
```

```python
# Employee List
employee= ['Abhijeet', 'Alice', 'Bob', 'Alice', 'Stuti', 'Ankur']
```

```python
#Check the Team Size
def evaluate_team(employee):
    if len(employee)>5:
        print("Team is Big") 
    else:
        print("Team is Small")

evaluate_team(employee)
 
```

```python
#Loop with index
for idx,member in enumerate(employee):
    
    print(f"Member {idx+1} ")
    greet(member)
    print("************")
```
