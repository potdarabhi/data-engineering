### Handling NULLs in DF

```python
display(df)
```

```python
df_modified = df.dropna(subset=['name','email'])
display(df_modified)
```

```python
df['department'] = df['department'].fillna('Unknown')
df['salary'] = df['salary'].fillna(df['salary'].mean())
display(df)
```

```python
df['join_date'] = pd.to_datetime(df['join_date'],errors='coerce') #convert invalid date to NaT
df['join_date'] = df['join_date'].fillna(method='ffill') #Fill na with forward fill
display(df)
```

### Handling Outliers

```python
print(df['salary'].describe())
```

```python
#removed outliers i.e., 200000 from salary as it was higher than 75%
df = df[df['salary']<200000]
display(df)
```
