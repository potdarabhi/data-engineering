### Series

```python
import pandas as pd
```

```python
salaries = pd.Series([50000,60000,70000,80000])
print(salaries)
```

```python
#creating custom index
salary = pd.Series([50000,60000,70000], index = ['Abhijeet','Omkar','Stuti'])
print(salary)

print(salary['Abhijeet'])
```

### Series with mixed type

```python
details = pd.Series({'Name':'Abhijeet','Age':25,'Salary':50000})
print(details)
```

```python
salaries = pd.Series([50000,60000,70000])
updates_salaries = salaries * 1.10
print(updates_salaries)
```

Handling missing values


```python
data = pd.Series([100,None,200])
print('Original series\n',data)
```

```python
#to check if data is null
print('Is Null\n',data.isnull())
```

```python
print('Data filled\n',data.fillna(150))
```

### DataFrames in Pandas



A DataFrame is a 2D labeled table of rows and columns. Think of it as an in-memory Excel sheet or a SQL table.

```python
data1 = {
    'Name':['Abhijeet','Omkar','Stuti'],
    'Salary':[50000,60000,70000],
    'Desiganation':['Data Engineer','Data Analyst','Data Scientist'],
    'Location':['Pune','Mumbai','Bangalore']
}

df = pd.DataFrame(data1)
display(df)
```

```python
#column level access
display(df['Salary'])
```

```python
#row level acccesss
print(df.loc[0])
```

```python
print(df.iloc[1])
#multiple columns
print(df[['Name','Salary']])
#multiple rows
print(df.loc[[0,1]])
print(df.iloc[[0,1]])
```

```python
df['Bonus'] = df['Salary'] * 0.10
display(df)
```
