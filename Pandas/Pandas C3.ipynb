```python
import pandas as pd
```

```python
df = pd.read_csv('sales_messy.csv')
display(df.head())
```

```python
print('Null count',df.isnull().sum())
```

```python
print('Data types\n',df.dtypes)
```

```python
display(df['product'])
```

```python
#Applying function to remove sapce and camel case to product

df['product'] = df['product'].str.strip().str.title()
display(df)
```

```python
print(df['product'].unique())
```

```python
print(df['amount'].describe())
```

```python
#fill the values with median in amount column
df['amount'] = df['amount'].fillna(df['amount'].median())
display(df)
```

```python
df['date'] = pd.to_datetime(df['date'],errors='coerce')
print('\n\nInavalid date:\n',df['date'].isnull().sum())

#fill the null dates with forward fill
df['date'] = df['date'].fillna(method='ffill')

display(df)
```

```python
#add some usefull columns for analytics

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
display(df)
```

### Final step to do a sanity check


```python
print('Info:',df.info())
print('\n\nDescribe:',df.describe())
print('\n\nHead:',df.head())
```
